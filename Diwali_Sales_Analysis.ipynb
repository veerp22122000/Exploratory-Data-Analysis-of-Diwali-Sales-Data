# Exploratory Data Analysis of Diwali Sales Data

### Analysis based on Gender

```python
# plotting chart on Gender to compare total purchases and amount spend by Men and women
fig,axes=plt.subplots(1,2)

ax=sns.countplot(x = 'Gender',data = df,hue='Gender',ax=axes[0])
for bars in ax.containers:
    ax.bar_label(bars)
    
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen,hue='Gender',ax=axes[1])
```

*From above graphs-----> most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
#plotting chart comparing total purchases and amount spent by people belonging to different Age Group and segregating data based on Gender

fig,axes=plt.subplots(1,2)

ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender',ax=axes[0])
for bars in ax.containers:
    ax.bar_label(bars)

sales_age = df.groupby(['Age Group','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age,hue='Gender',ax=axes[1])
```

*From above graphs------> most of the buyers are of age group between 26-35 yrs female*

### State

```python
 # plot graph for total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# graph to get the total amount spent from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

*From above graphs-----> most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra, Karnataka and Delhi respectively*


### Marital Status

```python
#plotting chart comparing total purchases and amount spent by married vs unmarried people and segregating based on Gender

fig,axes=plt.subplots(1,2)
ax = sns.countplot(data = df, x = 'Marital_Status',hue='Gender',ax=axes[0])

for bars in ax.containers:
    ax.bar_label(bars)

sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(12,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender',ax=axes[1])
```

*From above graphs----> most of the buyers are unmarried (women) and they have high purchasing power*

### Occupation

```python
#graph to get an idea about total orders by people working in different sector  
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#grap to get an idea about total amount spent by people working in different sector 
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

*From above graphs-----> most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
#graph for most purchased product category
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#graph for product category on most money is spent by customers
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

*From above graphs-----> most of the sold products are from Food, Clothing and Electronics category*

```python
#graph for most sold product 
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

## Conclusion:

### *Unmarried women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category.*
